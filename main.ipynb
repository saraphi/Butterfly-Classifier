### _Initializing Butterfly Classifier_

```python
from butterfly_classifier import *

src_training = "./data/training"
src_validation = "./data/validation"
src_test = "./data/test"

butterfly_classifier = ButterflyClassifier((224, 224), 15, src_training, src_validation, src_test)
```

### _Training_

```python
h = butterfly_classifier.train(epochs=100)
```

### _Training: Evaluation_

```python
import matplotlib.pyplot as plt

plt.plot(h.history['accuracy'])
plt.plot(h.history['val_accuracy'])
plt.plot(h.history['loss'])
plt.title('Model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation','loss'], loc='upper right')
plt.show()
```
